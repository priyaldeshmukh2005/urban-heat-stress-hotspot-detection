```python
import pandas as pd
import numpy as np
```

```python
sample_data = {
    "NDVI": [0.75, 0.42, 0.15, 0.67, 0.42],
    "NDBI": [0.20, 0.55, 0.80, 0.30, 0.65],
    "LST": [28.5, 35.2, 41.8, 30.4, 38.7],
    "PopulationDensity": [1200, 4500, 8000, 2000, 6000],
    "Humidity": [65, 55, 40, 70, 50],
    "WindSpeed": [12, 8, 5, 15, 7]
}

df = pd.DataFrame(sample_data)

df
```

```python
df["Green_Built_Ratio"] = (
    df["NDVI"] /
    (df["NDBI"] + 0.01)
)

df
```

```python
df["Cooling_Potential"] = (
    df["NDVI"] *
    df["Humidity"]
)

df
```

```python
df["Heat_Exposure_Index"] = (
    df["NDBI"] *
    df["PopulationDensity"]
)

df
```
